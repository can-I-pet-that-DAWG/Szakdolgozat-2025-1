```python
"""ezzel scrapelem a 10*5 szezont"""

def get_csv_links(base_url, file_prefix):
    response = requests.get(base_url)
    response.raise_for_status()
    
    soup = BeautifulSoup(response.text, 'html.parser')
    links = soup.find_all('a', href=True)
    
    csv_links = []
    for link in links:
        match = re.search(r'/([0-9]{4})/' + file_prefix + r'\.csv', link['href'])
        if match:
            season = match.group(1)
            full_link = link['href'] if link['href'].startswith('http') else f"https://www.football-data.co.uk/{link['href']}"
            csv_links.append((full_link, season))
    
    return csv_links[1:11] 

def download_csv_files(base_url, download_folder, file_prefix):
    os.makedirs(download_folder, exist_ok=True)
    csv_links = get_csv_links(base_url, file_prefix)
    
    for link, season in csv_links:
        file_name = os.path.join(download_folder, f"{season}.csv")
        response = requests.get(link)
        response.raise_for_status()
        
        with open(file_name, 'wb') as file:
            file.write(response.content)
        
    print(f"{file_name}")

if __name__ == "__main__":
    leagues = {
        "england": ("https://www.football-data.co.uk/englandm.php", "PremierLeague_CSVs", "E0"),
        "germany": ("https://www.football-data.co.uk/germanym.php", "Bundesliga_CSVs", "D1"),
        "italy": ("https://www.football-data.co.uk/italym.php", "SerieA_CSVs", "I1"),
        "spain": ("https://www.football-data.co.uk/spainm.php", "LaLiga_CSVs", "SP1"),
        "france": ("https://www.football-data.co.uk/francem.php", "Ligue1_CSVs", "F1")
    }
    
    for league, (url, folder, prefix) in leagues.items():
        download_csv_files(url, folder, prefix)
```

```python
def odds_data():

    """ez a másik forrás, innen jönnek a fogadással kapcsolatos statisztikák.
    egyébként ebben is vannak foci statisztikák, és teszteltem is random forestet
    ezekkel az adatokkal, de nem kaptam túl jó eredményeket.
    viszont van olyan változó amit érdemes lehet behozni, meg fogom nézni"""
    
    base_path = os.path.join(os.getcwd(), 'football-data')  
    league_folders = [f.path for f in os.scandir(base_path) if f.is_dir()]
    
    league_data = {}
    
    for league_folder in league_folders:
        league_name = os.path.basename(league_folder)
        csv_files = glob.glob(os.path.join(league_folder, '*.csv'))
    
        dataframes = [pd.read_csv(file) for file in csv_files]
    
        if dataframes:
            league_data[league_name] = pd.concat(dataframes, ignore_index=True)

    for key in league_data:  
        league_data[key].loc[:, "Avg_Over2.5"] = league_data[key]["BbAv>2.5"].fillna(league_data[key]["Avg>2.5"])
        league_data[key].loc[:, "Avg_Under2.5"] = league_data[key]["BbAv<2.5"].fillna(league_data[key]["Avg<2.5"])
        league_data[key].loc[:, "Avg_Home"] = league_data[key]["BbAvH"].fillna(league_data[key]["AvgH"])
        league_data[key].loc[:, "Avg_Away"] = league_data[key]["BbAvA"].fillna(league_data[key]["AvgA"])
        league_data[key].loc[:, "Avg_Draw"] = league_data[key]["BbAvD"].fillna(league_data[key]["AvgD"])

        
    keep = ['Date','HomeTeam','HS', 'AS','AST', 'HST', 'HTHG','HTAG', 'AwayTeam','FTHG','FTAG',
            "Avg_Over2.5",  "Avg_Under2.5","Avg_Home","Avg_Away","Avg_Draw"]    
    rename_dict = {
        "Avg_Over2.5": "O2.5_odds_market",
        "Avg_Under2.5": "U2.5_odds_market",
        "Avg_Home": "HomeW_odds_market",
        "Avg_Away": "AwayW_odds_market",
        "Avg_Draw": "Draw_odds_market"
    }
    for key in league_data:
        league_data[key] = league_data[key][keep].rename(columns=rename_dict)

    for key in league_data:
        league_data[key]['Date'] = league_data[key]['Date'].astype(str)
        league_data[key]['Date'] = pd.to_datetime(league_data[key]['Date'], dayfirst=True, errors='coerce')
    
    for league_name, df in league_data.items():
        short_name = league_name.split('_')[0] + "_odds"  
        globals()[short_name] = df  

    # with open("league_data.pkl", "wb") as f:
    #     pickle.dump(league_data, f)

    league_data = pd.concat(
        [df.assign(league_id=i + 1) for i, df in enumerate(league_data.values())], 
        ignore_index=True
    )
    
    id_mapping = {4: 1, 3:5, 1: 3, 5: 2, 2: 4}
    league_data['league_id'] = league_data['league_id'].replace(id_mapping)
    league_data.dropna(inplace=True)

    
    return league_data
```

```python
def create_dataframes(big5, league_data):
    
    """Létrehozzuk a kombinált ligák mergelt adatbázisát, ez kb 18000 sor,
    meccsbeli és fogadási statisztikákkal"""
    
    league_data['Date'] = pd.to_datetime(league_data['Date'])
    big5['date'] = pd.to_datetime(big5['date'])

    league_data = league_data.sort_values(['league_id', 'Date', 'HomeTeam', 'AwayTeam', 'FTHG', 'FTAG'])
    league_data.reset_index(drop=True, inplace=True)

    big5 = big5.sort_values(['league_id', 'date', 'home_team', 'away_team', 'home_goals', 'away_goals'])
    big5.reset_index(drop=True, inplace=True)

    merged = pd.merge(
        big5,
        league_data,
        left_on=['home_team', 'away_team'],
        right_on=['HomeTeam', 'AwayTeam'],
        how='outer',
        suffixes=('_big5', '_league')
    )

    def date_within_tolerance(row):
        return abs((row['date'] - row['Date']).days) <= 2
    merged['date_match'] = merged.apply(date_within_tolerance, axis=1)
    merged = merged[merged['date_match']]
    merged.drop(columns=['date_match'], inplace=True)
    merged.dropna(subset=['game_id'], inplace=True)
    merged = merged.sort_values(['league_id_big5', 'date', 'home_team', 'away_team']).reset_index(drop=True)

    match = [
        'game_id', 'league_id_big5', 'season', 'date', 'home_team', 'away_team',
        'away_goals', 'away_points', 'away_expected_points', 'away_xg',
        'away_np_xg', 'away_np_xg_difference', 'away_ppda',
        'away_deep_completions',
        'home_goals', 'home_points',
        'home_expected_points', 'home_xg', 'home_np_xg',
        'home_np_xg_difference', 'home_ppda', 'home_deep_completions',
        'HS', 'AS', 'AST', 'HST', 'HTHG', 'HTAG',
    ]

    odds = [
        'game_id', 'O2.5_odds_market', 'U2.5_odds_market',
        'HomeW_odds_market', 'AwayW_odds_market', 'Draw_odds_market'
    ]

    combined_odds = merged[odds].reset_index(drop=True)
    combined_match = merged[match].reset_index(drop=True)
    combined_match = combined_match.rename(columns={"league_id_big5": "league_id"})
    # merged.to_excel('understat_odds_14-24_big5.xlsx', index=False)

    return combined_match, combined_odds, merged
```

```python
def separate_matches(dictionary, columns_home, columns_away):

    """kombináljuk a _home és _away df-ket.
    végeredményben 1 mérkőzés 2 sor, amely sorokban mind az idegen, mind a hazai nézőpontból
    láthatók lesznek a statisztikák"""
    
    league_combined = {}

    for league, data in dictionary.items():
        df_home = data['home'].copy()
        df_away = data['away'].copy()
    
        df_home.sort_values(['game_id'], inplace=True)
        df_away.sort_values(['game_id'], inplace=True)
    
        df_home.reset_index(drop=True, inplace=True)
        df_away.reset_index(drop=True, inplace=True)
        
        df_home['goals_for'] = df_home['home_goals']
        df_home['goals_against'] = df_away['away_goals']
    
        df_away['goals_for'] = df_away['away_goals']
        df_away['goals_against'] = df_home['home_goals']

        df_home = df_home[['game_id', 'date', 'season', 'home_team', 'goals_for', 'goals_against'] + columns_home].rename(
            columns={'home_team': 'team', **{col: col.replace('home_', '') for col in columns_home}})
        df_home['is_home'] = 1  
                                #is_home dummy, enelkul nincs értelme a goals_for és goals_against változóknak,
                                #plusz a haza pálya előnye miatt információtartalma is van
                                #később látni fogod, hogy benne van az elemzésben, de kérdés, hogy jó-e, ha
                                #a modell megtanulja ezt is

        df_away = df_away[['game_id', 'date', 'season', 'away_team', 'goals_for', 'goals_against'] + columns_away].rename(
            columns={'away_team': 'team', **{col: col.replace('away_', '') for col in columns_away}})
        df_away['is_home'] = 0  

        df_combined = pd.concat([df_home, df_away], ignore_index=True).sort_values(by=['season', 'team', 'date'])
        df_combined['goals_diff'] = df_combined['goals_for'] - df_combined['goals_against']
        df_combined = df_combined.drop(columns='goals')
        league_combined[league] = df_combined

    return league_combined
```

```python
def streak_variables(dictionary):
    for league, df in dictionary.items():
        df = df.copy()
        df = df.sort_values(by=['season', 'team', 'date'])  

        df['is_win'] = (df['goals_for'] > df['goals_against']).astype(int)
        df['is_lose'] = (df['goals_for'] < df['goals_against']).astype(int)
        df['is_draw'] = (df['goals_for'] == df['goals_against']).astype(int)
        df['is_over_2_5'] = ((df['goals_for'] + df['goals_against']) > 2).astype(int)
        df['is_under_2_5'] = ((df['goals_for'] + df['goals_against']) <= 2).astype(int)

        for col, new_col in zip(
            ['is_win', 'is_lose', 'is_draw', 'is_over_2_5', 'is_under_2_5'], 
            ['last5_wins_count', 'last5_losses_count', 'last5_draws_count', 'last5_over_2_5_count', 'last5_under_2_5_count']
        ):
            df[new_col] = (
                df.groupby(['season', 'team'])[col]
                .rolling(window=5, min_periods=1)
                .sum()
                .shift(1)
                .reset_index(level=[0, 1], drop=True)
            )
        
        df['last5_total'] = df['last5_wins_count'] + df['last5_losses_count'] + df['last5_draws_count']
        df['last5_total'] = df['last5_total'].replace(0, 1)  

        df['last5_win_rate'] = df['last5_wins_count'] / df['last5_total']
        df['last5_lose_rate'] = df['last5_losses_count'] / df['last5_total']
        df['last5_draw_rate'] = df['last5_draws_count'] / df['last5_total']

        df.drop(columns=['last5_total'], inplace=True)
        
        df.loc[df.groupby(['season', 'team']).head(1).index, ['last5_wins_count',
                                                              'last5_losses_count',
                                                              'last5_draws_count',
                                                              'last5_over_2_5_count',
                                                              'last5_under_2_5_count',
                                                              'last5_win_rate',
                                                              'last5_lose_rate',
                                                              'last5_draw_rate']] = 0

        df.drop(columns=['is_win', 'is_lose', 'is_draw', 'is_over_2_5', 'is_under_2_5'], inplace=True)
        
        dictionary[league] = df

    return dictionary
```

```python
dict_with_odds = streak_variables(dict_with_odds)
```

```python
def rolling_avg_per_team(dictionary, columns, window=5):

    """itt számoljuk ki a mozgóátlagokat.
    az alapvető elmélet emögött, hogy a modell nem tanulhat a jövőből,
    ergo nem lehetnek a végső adathalmazban olyan változók, mint 'goals',
    azonban az egyes csapatok formáit meg lehet fogni ezekkel a mozgóátlagokkal.
    hogyan --> 1. csoportositunk season és csapat alapján
               2. dátum szerint rendezünk
               3. minden groupon végigloopolunk, úgy, hogy minden csapatnak kiszámoljuk az átlagát az előző i meccsre.
                  az első meccset minden szezonban üresen hagyjuk, mert akkor még nincs formája a csapatnak, és a modell nem
                  tanulhat a jövőből, tehát nem ismerheti a szezon első meccsén a teljesítményeket. utána addig várunk, amíg
                  i = megadott window size. ha ez 5, akkor az ötödik meccs az első, ami már 5 értékkel dolgozik. onnantól ez
                  az ablak mozog a szezon végéig, ahol resetel. lehet érdemes lenne súlyozni."""
    
    league_transformed = {}

    for league, df in dictionary.items():
        print(f"{league} kész")
        df = df.sort_values(by=['season', 'team', 'date']) 
        new_df = df.copy()

        for col in columns:
            new_df[f"{col}_RA"] = np.nan  

        grouped = new_df.groupby(['season', 'team']) 
        
        for (season, team), group in grouped:
            group = group.sort_values(by='date')  
            indexes = group.index  

            for col in columns:
                new_col_name = f"{col}_RA"
                values = group[col].tolist()  
                rolling_avgs = [np.nan]  
                
                for i in range(1, len(values)):  
                    if i < window:  
                        rolling_avgs.append(sum(values[:i]) / i)  
                    else:
                        # rolling_avgs.append(sum(values[i-window+1:i+1]) / window)  
                        rolling_avgs.append(sum(values[i-window:i]) / window)


                new_df.loc[indexes, new_col_name] = rolling_avgs  

        new_df = new_df.sort_values(by=['date', 'season'])  
        league_transformed[league] = new_df

    return league_transformed
```

```python
def cumulative_points(dictionary):
    for league, df in dictionary.items():
        df = df.sort_values(by=["team", "season", "date"])  

        df["points_cumulative"] = df.groupby(["season", "team"])["points"].cumsum()
        df["points_cumulative"] = df.groupby(['season', 'team'])["points_cumulative"].shift(fill_value=0)

        dictionary[league] = df.dropna()

    return dictionary
```

```python
def prepare_to_model(dictionary):

    """ebben a függvényben dátum és game_id szerint rendezek, 
    majd kidobok minden olyan változót, amit a modell nem tud értelmezni.
    az is_home részemről kérdéses, egyelőre bennehagytam.
    outputként kapok egy dictionaryt az összes ligával,
    és mindegyik ligát megkapom külön df-ként is"""
    
    prepare_to_model_dict = {}

    for league, df in dictionary.items():
        df = df.sort_values(by=['date', 'game_id']).reset_index(drop=True)
        
        df['target_o2.5'] = (df['goals_for'] + df['goals_against'] >= 3).astype(int)
        df['target_home_win'] = ((df['points'] == 3) & (df['is_home'] == 1) | (df['points'] == 0) & (df['is_home'] == 0)).astype(int)
        df['target_away_win'] = ((df['points'] == 3) & (df['is_home'] == 0) | (df['points'] == 0) & (df['is_home'] == 1)).astype(int)
        df['target_draw'] = (df['goals_for'] == df['goals_against']).astype(int)

        cols = [
            'date','game_id','target_o2.5', 'target_home_win', 'target_away_win', 'target_draw','is_home',
            'O2.5_odds_market', 'U2.5_odds_market', 'HomeW_odds_market',
            'AwayW_odds_market', 'Draw_odds_market', 'goals_for_RA', 'goals_against_RA','shots_RA', 'shots_OT_RA', 'halftime_goals_RA',
            'points_RA','expected_points_RA', 'xg_RA', 'np_xg_RA',
            'np_xg_difference_RA', 'ppda_RA', 'deep_completions_RA', 'points_cumulative','cumulative_goals_diff',
            'last5_wins_count', 'last5_losses_count','last5_draws_count', 'last5_over_2_5_count', 'last5_under_2_5_count',
            'last5_win_rate', 'last5_lose_rate', 'last5_draw_rate']

        
        transformed_df = df[cols].copy()
        prepare_to_model_dict[league] = transformed_df 
        globals()[league.lower().replace(' ', '_')] = transformed_df  
        transformed_df = transformed_df.sort_values(['date','game_id'])

    return prepare_to_model_dict
```

```python
def ready_to_model(dictionary):

    merged_leagues = {}

    for league, df in dictionary.items():
        df = df.sort_values(by=['date','game_id']).reset_index(drop=True)
        df_home = df[df["is_home"] == 1].copy()
        df_away = df[df["is_home"] == 0].copy()

        ra_cols = [col for col in df.columns if col.endswith("_RA")]
    
        keep_cols = ["points_cumulative",
                     'last5_win_rate',
                     'last5_lose_rate',
                     'last5_draw_rate',
                     'cumulative_goals_diff',
                     'last5_wins_count',
                     'last5_losses_count',
                     'last5_draws_count',
                     'last5_over_2_5_count',
                     'last5_under_2_5_count'] + ra_cols

        home_cols = {col: f"home_{col}" for col in keep_cols}
        away_cols = {col: f"away_{col}" for col in keep_cols}

        df_home.rename(columns=home_cols, inplace=True)
        df_away.rename(columns=away_cols, inplace=True)

        df_home.drop(columns=["is_home"], inplace=True)
        df_away.drop(columns=["is_home"], inplace=True)

        df_merged = df_home.merge(df_away, on="game_id", suffixes=("_x", "_y"))
        
        df_merged.drop(columns=[col for col in df_merged.columns if col.endswith("_y")], inplace=True)
        df_merged.rename(columns={col: col[:-2] for col in df_merged.columns if col.endswith("_x")}, inplace=True)

        # df_merged.drop(columns=["game_id", 'date'], inplace=True)
        df_merged.drop(columns=["game_id"], inplace=True)


        merged_leagues[league] = df_merged
        globals()[league.lower().replace(' ', '_')] = df_merged 

    return merged_leagues
```

```python
model_ready_dict=ready_to_model(prepare_to_model_dict)
```

```python
def additional_variables(dictionary):
    for league, df in dictionary.items():
        df = df.copy()
        
        df["home_clinicality"] = df["home_goals_for_RA"] / df["home_xg_RA"].replace(0, 1)
        df["away_clinicality"] = df["away_goals_for_RA"] / df["away_xg_RA"].replace(0, 1)
        
        df["home_total_goals_RA"] = df["home_goals_for_RA"] + df["home_goals_against_RA"]
        df["away_total_goals_RA"] = df["away_goals_for_RA"] + df["away_goals_against_RA"]
        
        df['win_rate_diff'] = df['home_last5_win_rate'] - df['away_last5_win_rate']
        df['lose_rate_diff'] = df['home_last5_lose_rate'] - df['away_last5_lose_rate']
        df['draw_rate_diff'] = df['home_last5_draw_rate'] - df['away_last5_draw_rate']
        df['cumulative_points_diff'] = df['home_points_cumulative'] - df['away_points_cumulative']
        
        df['GD_diff'] = df['home_cumulative_goals_diff'] - df['away_cumulative_goals_diff']
        
        df.drop(columns=['date'], inplace=True)

        dictionary[league] = df
        globals()[league.lower().replace(' ', '_')] = df  

    return dictionary
```
