```python
data = {
    'Premier League': [0, -1.75, 1, -1.14, -1.49],
    'Bundesliga': [-4.12, 0, 2.26, -1.47, -3.31],
    'Serie A': [-1.75, -5.79, 0, 0.03, -1.74],
    'La Liga': [-2.27, -1.8, 1.66, 0, -4.85],
    'Ligue 1': [-1.15, -2.35, 0.68, 1.12, 0]
}

index = ['Premier League', 'Bundesliga', 'Serie A', 'La Liga', 'Ligue 1']
df = pd.DataFrame(data, index=index)
df_t = df.T

x_limits = (-6, 6) 

fig1 = plt.figure(figsize=(12, 3))
gs1 = gridspec.GridSpec(1, 3)  

for i in range(3):
    ax = fig1.add_subplot(gs1[0, i])
    league = df_t.index[i]
    col = df_t.loc[league].drop(league)
    
    colors = ['green' if val > 0 else 'red' for val in col.values]
    
    col.plot(kind='barh', ax=ax, color=colors)
    ax.set_title(f'{league}')
    ax.axvline(0, color='gray', linestyle='--', linewidth=1)
    ax.set_ylabel('')
    ax.set_xlabel('')
    ax.grid(axis='x', linestyle='--', alpha=0.5)
    ax.set_xlim(x_limits)

plt.tight_layout()
plt.savefig('abrak/logreg_top_charts.jpg', format='jpg', dpi=300)
plt.show()

fig2 = plt.figure(figsize=(8, 3))
gs2 = gridspec.GridSpec(1, 2) 

for i in range(2):
    ax = fig2.add_subplot(gs2[0, i]) 
    league = df_t.index[i + 3]
    col = df_t.loc[league].drop(league)
    
    colors = ['green' if val > 0 else 'red' for val in col.values]
    
    col.plot(kind='barh', ax=ax, color=colors)
    ax.set_title(f'{league}')
    ax.axvline(0, color='gray', linestyle='--', linewidth=1)
    ax.set_ylabel('')
    ax.grid(axis='x', linestyle='--', alpha=0.5)
    ax.set_xlim(x_limits)

plt.tight_layout()
plt.savefig('abrak/logreg_bottom_charts.jpg', format='jpg', dpi=300)
plt.show()
```
